```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Timestamped data is the most basic type of time series data that associates values with points in time 

```python
data = {'date':[
                '2014-05-01 18:45:23.0293223',
                '2014-05-01 18:45:23.1093223',
                '2014-06-01 18:45:23.1193223',
                '2015-05-01 18:45:23.1293223',
                '2015-05-01 18:45:23.1493223',
                '2015-06-01 18:45:23.1593223',
                '2016-05-01 18:45:23.1693223',
                '2016-06-01 18:45:23.1793223',
                '2016-07-01 18:45:23.1893223',
                '2017-05-01 18:45:23.1993223'
               ], 
        'battle_d':
               [12,33,54,76,23,12,32,14,15,16]}
```

```python
data
```

```python
df = pd.DataFrame(data,columns=['date','battle_d'])
df.head()
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.index = df['date']
del df['date']
df
```

# Count # of observations per timestamp

```python
df.groupby(level=0).count()
```
