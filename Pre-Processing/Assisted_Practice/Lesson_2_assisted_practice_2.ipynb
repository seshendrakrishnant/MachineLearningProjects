```python
import pandas as pd
df = pd.readf = pd.read_csv("middle_tn_schools.csv")
df.describe()
```

# grouping the two columns by school rating 
# checking description of dataframe
# unstacking denormalizes the data

```python
#Phase 2
df[['reduced_lunch', 'school_rating']].groupby(['school_rating']).describe().unstack()
```

# Checking correlation between reduced_lunch and school_rating

```python
df[['reduced_lunch', 'school_rating']].corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlations = df.corr()
sns.heatmap(data=correlations,square=True,cmap='bwr')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
